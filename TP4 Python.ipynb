```python
from tkinter import *
from math import *


def table(value):
    global N
    N=int(value)
    show(N, P)

def vertex(value):
    global P
    P=int(value)
    show(N, P)

def show(n,p):
    cnv.delete(ALL)
    base=[(R*cos(2*k*pi/p), R*sin(2*k*pi/p)) for k in range(p)]
    draw_base(cnv, p, base, a, b)
    # Tracé de la table
    for k in range(p):
        segment(cnv, k, n, p, base, a, b)   


def passage(x,y, a, b):
    return (a+x, b-y)

def produit(k, n, p):
    return k*n % p

def segment(cnv, k, n, p, base, a, b):
    # base : les coordonnées des points du cercle
    # on trace le produit prod=k*n
    # k -> A
    # prod -> B
    prod=produit(k, n, p)
    xA, yA=base[k]
    xB, yB=base[prod]    
    A=passage(xA, yA, a, b)
    B=passage(xB, yB, a, b)
    cnv.create_line(A, B)
    
def dot(cnv, C, R=6, color='red'):
    xC, yC=C
    return cnv.create_oval(xC-R,yC-R,xC+R, yC+R, fill=color, outline=color)
  
def draw_base(cnv, p, base, a, b):
    # Les sommets sur le cercle
    for k in range(p):
        x, y=base[k]
        X, Y=passage(x, y, a, b)
        dot(cnv, (X, Y), R=2, color='black')
    

master=Tk()
a=b=1.2*R
cnv=Canvas(master, width=2*a, height=2*b, bg='ivory')
cnv.pack()
# Les points du cercle
base=[(R*cos(2*k*pi/p), R*sin(2*k*pi/p)) for k in range(p)]
draw_base(cnv, p, base, a, b)
# Tracé de la table
for k in range(p):
    segment(cnv, k, n, p, base, a, b)    

table_slider = Scale(master,label="Table de ",font="Arial 15 bold", orient = "horizontal", command= table, from_=2, to=50,length=250)
table_slider.pack(pady=15)
vertex_slider = Scale(master,label="Nombres de sommets  :",font="Arial 15 bold", orient = "horizontal", command= vertex, from_=2, to=500,length=250)
vertex_slider.pack(pady=5)
master.mainloop()
  



 
p=100 # nombre de points sur le cercle
R=200
n=5 # Table de n
N=P=2

    

```

```python
from tkinter import *

TOP=0
LEFT=0
RIGHT=0
BASE_H=TOP

SEP=8
N=6
P=7
RAD=20
DIAM=2*RAD

PERIOD=2*SEP+DIAM


BOARD_W=P*PERIOD
BOARD_H=N*PERIOD

WIDTH=LEFT+RIGHT+P*PERIOD
HEIGHT=BASE_H+N*PERIOD+TOP

def dot(cnv, C, R=6, color='red'):
    xC, yC=C
    return cnv.create_oval(xC-R,yC-R,xC+R, yC+R, fill=color, outline=color)

def play_token(cnv, line, col, color):
    x0=LEFT+0.5*PERIOD
    y0=TOP+0.5*PERIOD+(N-1)*PERIOD
    xC=x0+col*PERIOD
    yC=y0 - line*PERIOD 
    dot(cnv, (xC,yC), RAD, color=color)

def fill_board(cnv):
    x=x0=LEFT+0.5*PERIOD
    y=y0=TOP+0.5*PERIOD

    for i in range(P):
        for j in range(N):
            dot(cnv, (x,y), RAD, color='white')
            y+=PERIOD

        x+=PERIOD
        y=y0

    x=x0+3*PERIOD
    y=y0+(N-1)*PERIOD

def show():
    root = Tk()
    cnv = Canvas(root, width=WIDTH, height=HEIGHT, background="ivory")
    boutons=[]
    for i in range(7):
        btn=Button(root)
        boutons.append(btn)
        btn.grid(row=0, column=i, sticky="N"+"E"+"S"+"W")
    
    cnv.grid(row=1, columnspan=7)

    base=cnv.create_rectangle((0, HEIGHT), (WIDTH, 0), fill="blue", outline='')
    board=cnv.create_rectangle((LEFT, TOP), (WIDTH-RIGHT, HEIGHT-BASE_H), fill="blue", outline='')
    fill_board(cnv)
    
    play_token(cnv, 0, 3, 'yellow')        
    play_token(cnv, 0, 2, 'red') 
    play_token(cnv, 1, 2, 'yellow')        
    play_token(cnv, 2, 2, 'red')
    
    root.mainloop()
        

show()
```
