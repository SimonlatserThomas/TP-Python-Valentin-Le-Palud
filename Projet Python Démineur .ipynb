```python
from random import *
from tkinter import*
def afficher(M):
    "Affiche une matrice en respectant les alignements par colonnes"
    w=[max([len(str(M[i][j])) for i in range(len(M))]) for j in range(len(M[0]))]
    for i in range(len(M)):
        for j in range(len(M[0])):
            print("%*s" %(w[j],str(M[i][j])), end= ' ')
        print()    

def voisins(n, i, j):
    """Dans une grille n x n, indices (ligne, colonne) des voisins 
    dans la grille d'un sommet (i,j) de la grille"""
    return [(a,b) for (a, b) in [(i, j+1),(i, j-1), (i-1, j), (i+1,j)] if a in range(n) and b in range(n)]

def bomb(L,n,l,c):
    vbomb=voisins(n,l,c)#liste des voisions d'une bombe
    print(vbomb)
    for (i, j) in vbomb:
        x=L[i-1][j-1]
        if x!=9:
            x=x+1
        L[i-1][j-1]=x
        
L=[[],[],[],[],[],[],[],[]]
s=0
n=10
for i in range (len(L)):
    for j in range (8):
         L[i].append(0)

for p in range (11):
    l=randrange(8)
    c=randrange(8)
    L[l][c]=9
    bomb(L,n,l,c)
    
afficher(L)








WIDTH=500
HEIGHT=500
root=Tk()
unit=50
n=8

cnv = Canvas(root, width=unit*n, height=unit*n, background="ivory") 
cnv.pack(padx=10, pady=10, side=LEFT)



def jeu(event):   # quand on cliqe sur une case ça renvoit les coordonées correspondant à la liste L
    y=event.x//50
    x=event.y//50
    print(x,y)
    C=(x,y)
    if L[x][y]!=9:
        print("pas bombe")
    else:
        print("bombe !")
        


def init(): # initialise les case du jeu.
    for line in range(n):
        for col in range(n):
            A=(unit*col, unit*line)
            B=(unit*(col+1), unit*(line+1))
               
            cnv.create_rectangle(A, B, fill="grey",)
            

cnv.bind("<Button-1>",jeu)         
                
init()
root.mainloop()





















#Pour le clique pour afficher les case du démineur utiliser le propagate de l'exo du feu de forêt 
```
