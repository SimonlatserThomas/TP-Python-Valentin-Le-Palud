```python
 
from tkinter import *
from random import *

CANVAS_SIDE=400
NB_SQUARES=4
R=CANVAS_SIDE//NB_SQUARES



def fill_board(cnv, board):
    
    for i in range(NB_SQUARES):
        for j in range(NB_SQUARES):
            v=board[i][j]
            if not i==j==NB_SQUARES-1:
                cnv.create_rectangle(j*R, i*R, (1+j)*R, (1+i)*R, fill="lavender",tag="taquin")

                cnv.create_text(j*R+R//2, i*R+R//2, text=v,  font=('courier', CANVAS_SIDE//10, 'bold'),tag="taquin")
                
    
                                
def melanger(cnv):      
    cnv.delete("taquin")
    board=[[NB_SQUARES*lin+1+col for col in range(NB_SQUARES)] for lin in range(NB_SQUARES)]
    shuffle(board)
    for i in range(NB_SQUARES):
        for j in range(NB_SQUARES):
            
            v=board[i][j]
            if not i==j==NB_SQUARES-1:
                cnv.create_rectangle(j*R, i*R, (1+j)*R, (1+i)*R, fill="lavender",tag="taquin")
                cnv.create_text(j*R+R//2, i*R+R//2, text=v,  font=('courier', CANVAS_SIDE//10, 'bold'),tag="taquin")
    
    
def draw():
    master=Tk()
    cnv=Canvas(master, width=CANVAS_SIDE, height=CANVAS_SIDE, bg='ivory')
    cnv.pack()
    board=[[NB_SQUARES*lin+1+col for col in range(NB_SQUARES)] for lin in range(NB_SQUARES)]
    fill_board(cnv, board)
     
    bouton=Button(master, text="m√©langer",command=lambda: melanger(cnv))
    bouton.pack()
    
    master.mainloop()
    
draw()
```

```python
from random import *
from tkinter import *

def random_forest(p, n):
    units=[(line,col) for col in range(n) for line in range(n)]
    ntrees=int(n**2*p)
    trees=sample(units,ntrees)
    states=[[0]*n for _ in range(n)]
    for (i,j) in trees:
        states[i][j]=1    
    return states
                
def fill(cnv, states, unit):
    n=len(states)
    for line in range(n):
        for col in range(n):
            A=(unit*col, unit*line)
            B=(unit*(col+1), unit*(line+1))
            state=states[line][col]
            color=["ivory", "lime green", "red", "gray75"][state]
            id_=cnv.create_rectangle(A, B, fill=color, outline='',tag="green")
            ids.append(id_)
def fire(cnv):
    y=randrange(50)
    x=randrange(50)
    A=(unit*x, unit*y)
    B=(unit*x+10, unit*y+10)
    
    cnv.create_rectangle(A, B, fill="red", outline='')

def d (cnv,x):
    x=int(x)  
    cnv.delete(old)
    states=random_forest(x,n)
    fill(cnv, states, unit)
    

    
    
p=0.2

n=50

root = Tk()
unit=10
side=unit*n
cnv = Canvas(root, width=unit*n, height=unit*n, background="ivory")
cnv.pack(padx=10, pady=10)
states=random_forest(p,n)
fill(cnv, states, unit)

nbr=(((unit*n)**2)*p)/100
lbl=Label(root, text=nbr, font='Arial 30 bold')
lbl.pack(padx=15, pady=15)

bouton=Button(root, text="Fire !",command=lambda: fire(cnv))
bouton.pack()

curseur = Scale(root, orient = "horizontal",command=lambda: d(cnv,x), from_=0, to=100)
curseur.pack()


root.mainloop()
```
