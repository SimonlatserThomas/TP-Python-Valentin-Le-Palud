```python
from random import sample, randrange
from tkinter import *

COLORS=["ivory", "lime green", "red", "gray75"]

def random_forest(p, n):
    units=[(line,col) for col in range(n) for line in range(n)]
    ntrees=int(n**2*p)
    trees=sample(units,ntrees)
    states=[[0]*n for _ in range(n)]
    for (i,j) in trees:
        states[i][j]=1    
    return states, trees

def voisins(n, i, j):
    """Dans une grille n x n, indices (ligne, colonne) des voisins 
    dans la grille d'un sommet (i,j) de la grille"""
    return [(a,b) for (a, b) in [(i, j+1),(i, j-1), (i-1, j), (i+1,j)] if a in range(n) and b in range(n)]

def update_states(states, fires):
    n=len(states)
    to_fire=[]
    for (line, col) in fires:
        for (i, j) in voisins(n, line, col):
            if states[i][j]==1:
                to_fire.append((i, j))
    for (line, col) in to_fire:
        if states[line][col]==1:
            states[line][col]=2
    for (line, col) in fires:
        states[line][col]=3
    
    return list(set(to_fire)) 
                
def fill(cnv, states, unit):
    n=len(states)    
    for line in range(n):
        for col in range(n):
            A=(unit*col, unit*line)
            B=(unit*(col+1), unit*(line+1))
            state=states[line][col]
            color=COLORS[state]
            cnv.create_rectangle(A, B, fill=color, outline='')

def start_fire(states, trees):
    "Met le feu aléatoirement (code = 2) à l'arbre en position (i,j)"
    i,j= trees[randrange(len(trees))]
    states[i][j]=2
    return (i,j)

def new_forest():
    global ntrees, nfires, states, fires, trees, anim, pause
    print(anim)

    nfires=0
    states, trees=random_forest(p, n)
    ntrees=len(trees)
    cnv.delete("all")

    fill(cnv, states, unit)
    anim="new"
    pause=False
    lbl["text"]="%3s %%" %0
    stop_btn["text"]="PAUSE"

    
def go():
    global anim
    print(anim)
    
    if anim!="new" or pause==True:
        return
    global fires, cpt
    cpt=0
    i, j=start_fire(states, trees)
    fires=[(i,j)]  
    fill(cnv, states, unit)
    cnv.delete("all")
    anim="running"
    propagate()
    
def stop():
    global pause, anim
    print(anim)
    
    if anim == "new" and not pause:
        return
    if pause:
        anim="running"
        propagate()
        stop_btn["text"]="PAUSE"
        pause=False
    else:
        anim="new"
        stop_btn["text"]="BURN"    
        pause=True
    
def propagate():
    global nfires, fires, cpt, anim
    if anim=="new":
        return

    fires=update_states(states, fires)
    k=len(fires)
    nfires+=k
    cnv.delete("all")
    fill(cnv, states, unit)
    cnv.after(200, propagate)
    lbl["text"]="%3s %%" %(int(nfires/ntrees*100))    
    if k==0:
        
        return
    print(cpt)
    cpt+=1
   
def set_density(val):

    global p
    if anim=="running" or pause:
        return
    p=float(val)/100
    new_forest()

p=0.60
n=50
unit=10
cpt=0
anim=None

pause=False

root = Tk()
side=unit*n


cnv = Canvas(root, width=unit*n, height=unit*n, background="ivory") 
cnv.pack(padx=10, pady=10, side=LEFT) 

btn=Button(root,text="Nouvelle\nforêt",  font='Arial 20 bold', bg="lime green", command=new_forest,  width=8)
btn.pack(padx=10, pady=10)

# Bouton mettre le feu
go_btn=Button(root,text="Allumer\nle feu !",  font='Arial 20 bold', bg="pink", command=go, width=8)
go_btn.pack()

stop_btn=Button(root,text="PAUSE",  font='Arial 20 bold', command=stop, width=8)
stop_btn.pack(padx=10, pady=10)

lbl=Label(root,text="%3s %%" %0,  font='Arial 25 bold', bg='pink', width=5)
lbl.pack(padx=10, pady=10, side=LEFT)

    
curseur = Scale(root, orient = "vertical", command=set_density, from_=0, to=100, length=200) 
curseur.pack(padx=50)
curseur.set(100*p) 

new_forest()
root.mainloop()
```
